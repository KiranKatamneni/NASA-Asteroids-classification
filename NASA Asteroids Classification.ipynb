```python
for col in data.columns:
    print(f'{col}:   {data[col].nunique()}')
```

```python
data.drop(['Orbiting Body','Equinox'],axis = 1,inplace = True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
data['harzardous'] = enc.fit_transform(data.Hazardous)
```

```python
data.drop('Hazardous',axis = 1,inplace = True)
```

```python
data.head()
```

```python
X = data.drop('harzardous',axis = 1)
y = data.harzardous
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
pred1 = clf.predict(X_test) 
```

```python
accuracy_score(y_test,pred1)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize = (20,10))
tree.plot_tree(clf,max_depth=2,filled = True,fontsize = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
pred2 = rf.predict(X_test)
```

```python
accuracy_score(y_test,pred2)
```
